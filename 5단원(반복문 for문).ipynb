## **반복문이란?**
* 단순히 같은 작업을 여러 번 반복해야 할 때, 반복문을 활용하면 시간 소모를 줄일 수 있습니다. 
* 반복문은 같은 코드를 여러 번 입력하지 않고도 여러 번 반복하여 실행시킬 수 있는 방법입니다.
* 기본적으로 for를 사용한 반복문과 while을 사용한 반복문이 사용됩니다.

#### **for문은 리스트 혹은 문자열과 함께 주로 사용됩니다.**


- 설명 : 리스트의 첫번째 요소부터 마지막 요소까지 변수에 차례대로 대입해 코드1을 실행합니다. 
            

```python
# 예제 1 : list와 함께 사용하는 for문
for i in [1,2,3]:
    print(i)
```

#### **예제 1 해설**
i는 리스트 [1,2,3,4] 내부의 요소들로 값이 바뀌어집니다.   
첫번째 반복에서 i는 1, 두번째 반복에서는 2, 세번째 반복에서는 3이 됩니다. 

```python
# 예제 2 : 문자열과 함께 사용하는 for문
for i in 'Codle':
    print(i)
```

#### **예제 2 해설**
i는 문자열 'Codle' 내부의 요소들로 값이 바뀌어집니다.     
첫번째 반복부터 c,o,d,l,e 순서로 i는 바뀌게 됩니다.

> **반복문 활용 예시**

이럴 때, 반복문을 활용한다면, 코드를 여러 번 붙여넣지 않고도 문제를 해결할 수 있습니다.  

### range 함수 활용하기 
range() 함수는 아래와 같은 방법으로 사용됩니다. 

> **range(시작, 끝, 변화량)**  
시작점 이상 끝점 미만의 범위에서, 변화량 만큼 변화 시킵니다.  
예를 들어 range(0,10,1) 코드는 0,1,2,...,9 를 나타냅니다.(끝점은 포함되지 않는다는 점에 주의해야 합니다.)   
마찬가지로 range(0,10,2) 코드는 0, 2, 4, 6, 8 을 나타냅니다.(변화량이 2이기 때문에 2씩 변화하는 모습을 볼 수 있습니다.)   
시작점이 0이라면, range(10, 1)과 같이 시작점을 생략할 수 있습니다. 이 경우 range(10,1)은 range(0,10,1)은 같은 범위를 나타냅니다.   
변화량이 1이라면, range(0,10)으로 표현해 변화량을 생략할 수 있습니다. 따라서 range(0,10,1)은 range(10)으로도 표현할 수 있습니다.   
또한 변화량이 음수여도 상관 없습니다. 그럴 때는 시작점이 끝점보다 큽니다.  

```python
#예제 4 : Codle 100번 출력하기
for i in range(100):  # i = 0~99
    print("Codle")
```

```python
#예제 5 : 변화량이 음수인 경우
for i in range(4,0,-1):  
    print(i)
```

### **반복문의 활용 : 리스트와 index 값을 활용하기** 
> 리스트에서 각 값들은 인덱스를 가지고 있다고 했었습니다. 이를 활용해 봅시다. 


```python
#예제 6: 리스트와 index 값을 활용하기
bts = ["RM","진","슈가","지민","정국","뷔","제이홉"]

for i in range(7):
    print("방탄소년단의 "+str(i+1)+"번째 멤버는 "+ bts[i]+"입니다.")  # 인덱스는 0부터, 순서는 1부터 시작하기 때문에, str(i+1), bts[i] 와 같이 작성됩니다.
```

위 리스트 bts는 총 7개의 값을 가지고 있어 이들의 index는 0-6까지입니다.  
i는 그래서 0-6까지의 값으로 설정해두고 위와 같이 출력할 수 있습니다. 

### **반복문의 활용 : if 문과 함께 활용하기** 
> for문 속 변수에 if문을 적용해봅시다.     
for문안에 if문을 넣을 때는 반드시 tab키(혹은 스페이스바 4번)를 눌러 들여쓰기(indentation)를 해주어야 합니다.

```python
#예제 7: 1~i까지 3의 배수만 더하기

n = input("정수를 입력하세요.") # 변수 n에 정수를 입력받자.
n = int(n) # 입력받은 n을 정수형으로 변환해준다.
sum = 0 # 총합을 저장할 변수 sum을 0으로 설정하자.

# 0에서 n까지 순서대로 i에 대입하고, i가 3의 배수일 때에만 sum에 더해주는 반복문입니다. 
for i in range(n+1): # range 함수는 n미만까지가 범위이므로, n+1까지 범위를 설정해주자. 
    if i % 3 == 0: # 3으로 나눈 나머지가 0이면, 3의 배수라는 뜻이죠? (% : 나머지 연산자)
        sum = sum+i  #  i가 3의 배수일 때에만(조건 성립 시에만) sum에 i를 더해줍니다.
        
     
print(sum)
```

### **이중 for문 활용하기** 
> for문 속에 for문을 넣으면 어떤 일이 일어날까요?  
for문 안에 또 다른 for문을 중첩해서 작성하는 코드를 "이중 for문"이라고 합니다.   
이중 for문은 아래와 같은 구조로 이루어집니다. 

```python
for i in range(4):
    for j in range(3):
        print("i=", i, ", j=", j)
    print("=============")
```

> for i in range(4) 와 같이 바깥에 있는 for문과 for j in range(3) 과 같이 안 쪽에 있는 for문으로 이루어져 있습니다.   
이중 for문의 수행 과정은 다음과 같습니다.   
바깥쪽 for문이 실행되어 데이터 하나를 i에 저장하면, 그때 안쪽 for문이 실행됩니다.   
안쪽 for문의 반복이 모두 끝날 때까지 i는 처음 저장된 그 데이터를 유지하고,   
안쪽 for문 반복이 끝난 후에야 바깥쪽 for문의 반복이 다시 수행되어 i의 데이터가 바뀝니다.   
실제로 위에 제시된 이중 for문을 보면 i는 j가 range(3)만큼의 반복을 모두 끝낼 때까지 변화하지 않습니다.     
그리고 i가 range(4)만큼의 반복을 하여 마지막 데이터인 3을 저장하고 있을 때,  
j가 반복을 모두 끝내면 이중 for문의 반복이 종료됩니다.   

```python
#예제 8 :이중 for문 활용 - 위의 그림과 같은 패턴을 이중 for문을 활용하여 그려보세요.
'''
이중 for문은 패턴 그리기 문제에 많이 활용됩니다. 
2차원 상의 패턴은 행과 열로 표현되는데, 바깥쪽 for문이 행, 안쪽 for문이 열을 나타내도록 코드를 구성하면 패턴을 쉽게 구현할 수 있습니다.
(힌트) 위의 그림은 행이 10줄, 열은 10개의 숫자로 이루어져 있고, 각각은 바깥 for문과 안쪽 for문으로 구현됨에 유의하여 예제를 풀어보세요.

(참고) 1단원에서 말했지만, print() 함수는 마지막에 자동으로 줄바꿈을 수행합니다. 
줄바꿈을 하지 않게 하기 위해서는 출력의 마지막 문자를 바꿔야 합니다. 
그 방법은 print("출력할 것", end=" ") 와 같이 end를 다른 문자로 지정해주면 됩니다.
'''

for i in range(10):
    for j in range(1, 11):
        print(j, end=" ")
    print()
```

### **예제 8 해설**  
먼저, 바깥 쪽 for문에서 i=0을 됩니다. 이후 안쪽 for문을 반복시켜 1부터 10을 먼저 출력합니다.    
출력이 완료되면, print()를 하여 줄을 바꿔주고 바깥쪽 for문에서 range함수로 인해 i = 1로 변화하게 됩니다.  
그렇게 1~10이 총 10번 출력되게 됩니다.
