## **모듈이란?**
* 코딩을 활용해 새로운 프로그램을 만들기 위해서는 많은 코드가 필요합니다!
* 그럴 때마다 코드를 매번 새로 작성하기는 어렵기 때문에 자주 사용하는 함수들은 미리 모아놓을 수 있습니다.
* 이를 `모듈`이라고 합니다. 

예를 들어 하나의 모듈 `module01.py`는 변수 2개, 클래스 1개, 함수 2개를 포함하고 있습니다.   
이 때 main.py에서 `import module01`을 통해 모듈을 호출해주면 main.py에서 module01의 변수, 클래스, 함수를 모두 사용할 수 있습니다. 

## 그전에 참고할 것!  시퀀스는 무엇인가?
>시퀀스는 파이썬에서 연속형 자료형을 뜻합니다!  
우리가 배운 것 중에서는 list가 시퀀스에 속하죠.  
파이썬에서 시퀀스는 크게 list와 tuple, 이 두 가지가 있습니다.  
tuple은 내부 데이터 수정이 불가능하고, list는 내부 데이터를 수정할 수 있다는 특징이 있습니다.   
tuple은 수정이 불가능한 대신 메모리 효율성이 좋다는 장점이 있습니다.    
튜플은 (1,2,3,4,5)와 같이 소괄호를 사용해 데이터를 표기합니다. 
튜플도 리스트와 같이 인덱스를 활용할 수 있습니다. 




## **모듈 사용법과 대표적인 모듈**
### **모듈 사용법**
import 모듈명  
* `모듈명.함수`, `모듈명.변수`로 모듈 내의 함수와 변수를 사용할 수 있습니다.
  
import 모듈명 as 새이름  
* 모듈명을 새이름으로 사용할 수 있게 해줍니다. 
* 본래의 모듈명이 길거나 다른 이름과 혼동될 때, 다른 이름으로 바꿔서 사용합니다. 
   
from 모듈명 import 변수명(함수명)  
* import한 변수(함수)를 모듈명 없이 사용할 수 있습니다.  
  
from 모듈명 import *    
* 이 경우 모듈에 포함된 모든 변수(함수)를 모듈명 없이 사용할 수 있습니다. 

### **대표적인 모듈**
* import random 
>import random을 해야 사용할 수 있는 함수들입니다.
random.random() 이런식으로 random 뒤에 붙여서 활용합니다.
  
함수| 설명 
---|:---:
random() | 0~1 사이의 난수를 생성
randint(m,n) | m이상 n이하의 난수를 생성
choice(seq)|시퀀스의 요소들 중 하나를 무작위로 선택함.
shuffle(seq)|시퀀스의 요소들을 무작위로 섞음
sample(seq,i)|시퀀스의 요소 중에서 아무거나 i개를 반환하는 함수 

```python
#예제 1 : 1부터 10사이의 숫자를 무작위로 1개 선택하기, 실행할 때마다 n은 변화합니다.
import random

n = random.randint(1,10)
print(n)
```

```python
#예제 2 : list 무작위로 섞기
import random

a = [1,2,3,4,5,6]
random.shuffle(a)
print(a)
```

```python
#예제 3  random.shuffle()을 활용해서 로또 추첨기 만들기
import random 
lotto = [x+1 for x in range(45)] #1~45까지의 숫자를 저장합니다.
random.shuffle(lotto)
print(lotto[:6]) #list lotto의 앞 6자리를 출력합니다. 
```

```python
#예제 3-1 random.sample()을 활용해서 로또 추첨기 만들기

lotto = [x+1 for x in range(45)]
winner = random.sample(lotto,6)
print(winner)
```

## **구구단 게임을 만들어봅시다!**
1. a와 b는 1부터 9사이의 무작위 값으로 설정됩니다.  
2. "a*b는 무엇일까요?"가 출력되면, 정답을 입력해주세요!
3. 곱셈 정답이 맞으면, 점수가 1점 추가되고 다음 문제가 출제됩니다.
4. 틀리면, 최종 점수가 출력되고 게임이 종료됩니다. 

```python
#예제 4 구구단 게임 만들기 
import random
score = 0  # 점수를 저장할 변수입니다. 맞출 때마다 1씩 더해줄 겁니다. 
while True: # 무한 반복문입니다. (break로 종료시킬 수 있겠죠?)
    a = random.randint(1,9)
    b = random.randint(1,9)  # a, b에 모두 1~9사이의 정수를 무작위로 저장합니다.
    print(str(a)+"*"+str(b)+"는 무엇일까요?")

    answer = int(input())  # 정답을 입력받는 부분입니다. 
    if answer == a*b:
        score +=1  # 입력한 정답과 실제 a*b가 같으면 점수를 1점 더하고, 아래의 문구를 출력합니다. 
        print("정답입니다! 현재 점수 : "+ str(score))
        continue  # 뒤의 코드를 실행하지 않고 while문 조건으로 돌아갑니다. (사실 뒤의 코드는 else문이기 때문에 continue가 없어도 실행되지 않긴 합니다.)
    else :
        print("Game Over 최종 점수: "+ str(score))  # 입력한 정답과 실제 a*b가 다른 경우 점수를 출력한 뒤, break로 반복문을 종료합니다. 
        break
```

## **청개구리 가위바위보 게임을 만들어봅시다!**
1. 랜덤으로 가위, 바위,보 중 하나가 player와 computer에게 주어집니다. 
2. player가 computer를 이기면, 청개구리 가위바위보이기 때문에 반대로 "졌다"라고 입력해 주세요!  
3. 반대로, player가 computer에게 지면 "이겼다"라고 입력해주세요!  
4. 비기는 경우, "개굴"을 입력합니다.
5. 만약에 잘못 입력했다면 그 즉시 "Game Over"를 출력하고 반복문을 종료합니다. 잘못 입력한 경우 이외에는 게임을 계속 진행합니다!



```python
#예제 5 청개구리 가위바위보 게임
import random
kind = ["가위","바위","보"]  # 컴퓨터와 플레이어에게 각각 이 세 가지 중 하나를 랜덤하게 배정해야 합니다. 

while True:
    computer = random.choice(kind)  # 시퀀스, 이 경우 kind 리스트의 요소 중 하나를 랜덤하게 선택합니다. 
    player = random.choice(kind)
    print("플레이어 : ",player,",컴퓨터 : ", computer)
    talk = input("무엇이라 말해야할까요?")  # 정답을 입력받는 부분입니다. 
    
    if (player == "가위" and computer == "바위") or (player == "바위" and computer == "보") or (player == "보" and computer == "가위"): # player가 진 경우
        if talk == "이겼다": # 맞춘 경우
            continue
        else: # 틀린 경우
            print("Game Over")
            break  # 틀린 경우에는 반복문을 종료해줍니다. 
    elif (player == "바위" and computer == "가위") or (player == "보" and computer == "바위") or (player == "가위" and computer == "보"): # player가 이긴 경우
        if talk == "졌다": # 맞춘 경우
            continue
        else: # 틀린 경우
            print("Game Over")
            break
    elif (player == "바위" and computer == "바위") or (player == "보" and computer == "보") or (player == "가위" and computer == "가위"):  # 비긴 경우
        if talk == "개굴":  # 맞춘 경우
            continue
        else:  # 틀린 경우
            print("Game Over")
            break
    
```
