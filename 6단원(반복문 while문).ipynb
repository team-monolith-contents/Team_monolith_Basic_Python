## **반복문이란?**
* 같은 작업을 단순히 반복하는 업무를 해야 한다면, 시간 소모를 줄이기 위해 반복문을 통하여 해결할 수 있습니다.
* 반복문은 같은 코드를 여러번 입력하지 않고도 여러번 실행시킬 수 있는 방법입니다.
* 기본적으로 for를 사용한 반복문과 while을 사용한 반복문이 사용됩니다.

## - 형식 :  
   ### &nbsp;&nbsp;while 조건 :  
   ### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;명령어
- 설명 : while 다음의 조건이 참일 동안, 명령어 코드를 실행합니다. 
            

```python
# 예제 1 : while문 예씨
i =0 
while i<10:
    print("Hello, world!")
    i+=1
```

#### **조금 더 자세히 살펴볼까요?**
* i=0으로 설정된 후 while문으로 들어가 조건(i<10)을 만족하는지 살펴봅니다. i는 0이어서, 10보다 작으므로 조건을 만족합니다. 
* "Hello, world!"를 출력합니다. 
* i를 1 증가합니다.

그렇다면, 1부터 10까지 커질때까지 Hello, world!를 10번 출력하겠죠?

#### **초기식, 변화식, 조건**
* 초기식과 변화식은 while문을 종료시키기 위한 식들입니다.  
* 예를 들어 변화식이 없다면, i는 계속 0이기 때문에 while문이 종료하지 않고 무한하게 반복됩니다.    
* 그러나 무한하게 반복되는 것은 현실적으로 불가능하기 때문에, 이 경우 오류가 발생하게 됩니다.   
* 이러한 오류를 방지하기 위해서는 while문을 안정적으로 종료시켜주어야 하는데, while문의 종료는 초기식과 변화식으로 실현됩니다.   


### **실생활 예시 : 구구단을 for문과 while문으로 출력해봅시다.** 
사용자가 숫자를 입력하면, 입력한 단을 1에서 9까지 보여주는 구구단을 출력합니다. 

```python
# 예제 2 : for문으로 구구단 출력하기 
n = int(input("몇 단을 출력해볼까요?"))
for i in range(1,10):
    print(n,"*",i,"=",n*i)  # n은 고정되고, i만 1~9까지 반복하면 됩니다.
```

```python
# 예제 3 : while문으로 구구단 출력하기 
n = int(input("몇 단을 출력해볼까요?"))
i= 1
while i<10:
    print(n,"*",i,"=",n*i)
    i +=1  # i를 1~9까지 점점 커지도록 해줍니다.  
```

### **실생활 예시 : 반 평균 수학 점수 구하기**
반 학생 수를 입력 받고, 학생들의 수학 점수를 모두 입력 받아 평균값을 출력해봅시다.  

```python
#예제 4 : for 문을 활용해 반 평균 수학 점수 구하기
n = int(input("학생수를 입력하세요!"))
sum = 0  # "평균 = 수학 점수 총합 / 반 학생수"에서 수학 점수 총합을 구하기 위해 변수 sum을 선언해줍니다. (평균 = sum/n)
for i in range(n):  # 학생수만큼 반복해줍니다 
    score = int(input("학생 "+ str(i+1) + "의 점수를 입력하세요"))  # range(n)은 0부터 시작하기 때문에 학생번호를 나타내기 위해서는 +1을 해주어야 합니다. 
    sum = sum + score 

print("평균 점수 : ", sum/n)  
```

```python
#예제 5 : while 문을 활용해 반 평균 수학 점수 구하기
n = int(input("학생수를 입력하세요!"))
i = 0
sum = 0
while i<n:
    score = int(input("학생 "+ str(i+1) + "의 점수를 입력하세요"))  # 초기값을 0으로 설정했기 때문에 학생번호를 표현할 때 +1을 했습니다. (초기값을 1로 설정하는 방법도 있겠죠?)
    sum = sum + score 
    i = i+1
    
print("평균 점수 : ", sum/n)
```

### **반복문을 중간에 멈추고 싶다면? break를 활용하자!**
반복문으로 코드를 실행하던 중에 특정 조건에서 실행을 멈추거나 아예 종료하고 싶을 때가 있겠죠?  
이 때 break를 사용합니다! break가 있다면, 반복문의 조건과 상관없이 강제로 반복문을 중단시킵니다.  
입력한 값이 3의 배수이면 프로그램을 종료해보는 프로그램을 구현해봅시다!

```python
#예제 6: break 활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break  # n%3==0 일 때 반복문을 종료합니다. 따라서 다음의 "n은 3의배수가 아닙니다!"은 출력되지 않습니다. 
    print("n은 3의배수가 아닙니다!")
```

### **반복문을 중간까지만 실행하고 다시 처음으로 돌아오고 싶다면? continue를 활용하자!**
continue 또한 break와 마찬가지로 반복문 내에서 사용됩니다.  
break는 반복문을 강제로 중단한다면 continue는 반복문의 시작 부분으로 돌아가, for문의 변수를 진행하거나 while문의 반복조건을 판단합니다.  
즉 반복 중에 continue를 만나면, 해당 반복의 continue 이후의 코드는 실행하지 않고, 다시 반복문의 조건 부분으로 돌아가서 조건이 부합하면 다음 반복을 수행하게 됩니다.   
이번에는 입력한 값이 3의 배수이면 종료하고, 4의 배수인 경우 4의 배수입니다를 출력하고, 3의배수가 아닙니다!를 출력하지 않게 구현해봅시다!

```python
#예제 6: continue 활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break
    if n % 4 ==0:
        print("4의 배수가 입력되었습니다! 처음으로 돌아갑니다.")
        continue  # n%4==0 일 때 이후의 코드를 반복하지 않고, while문의 조건 부분으로 돌아갑니다. 
    print("n은 3의배수가 아닙니다!")
```
