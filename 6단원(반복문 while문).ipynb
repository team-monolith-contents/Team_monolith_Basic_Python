## **반복문이란?**
* 같은 작업을 단순히 반복하는 업무를 해야 한다면, 시간 소모를 줄이기 위해 반복문을 통하여 해결할 수 있습니다.
* 반복문은 같은 코드를 여러번 입력하지 않고도 여러번 실행시킬 수 있는 방법입니다.
* 기본적으로 for를 사용한 반복문과 while을 사용한 반복문이 사용됩니다.

## - 형식 :  
   ### &nbsp;&nbsp;while 조건 :  
   ### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;명령어
- 설명 : while 다음의 조건을 만족하는 동안, 명령어 코드를 실행합니다. 
            

```python
# 예제 1 : while문 예씨
i =0 
while i<10:
    print("Hello, world!")
    i+=1
```

#### **조금 더 자세히 살펴볼까요?**
* i=0으로 설정된 후 while문으로 들어가 조건을 만족하는지 살펴봅니다. i는 0이어서, 10보다 작으므로 조건을 만족합니다. 
* "Hello, world!"를 출력합니다. 
* i를 1 증가합니다.

그렇다면, 1부터 10까지 커질때까지 Hello, world!를 10번 출력하겠죠?

### **실생활 예시 : 구구단을 for문과 while문으로 출력해봅시다.** 
사용자가 숫자를 입력하면, 입력한 단을 보여주는 구구단을 출력합니다. 

```python
# 예제 2 : for문으로 구구단 출력하기 
n = int(input("몇 단을 출력해볼까요?"))
for i in range(1,10):
    print(n,"*",i,"=",n*i)
```

```python
# 예제 3 : while문으로 구구단 출력하기 
n = int(input("몇 단을 출력해볼까요?"))
i= 1
while i<10:
    print(n,"*",i,"=",n*i)
    i +=1
```

### **실생활 예시 : 반 평균 수학 점수 구하기**
반 학생 수를 입력 받고, 각각의 수학 점수를 입력받아 평균값을 출력해봅시다. 

```python
#예제 4 : for 문을 활용해 반 평균 수학 점수 구하기
n = int(input("학생수를 입력하세요!"))
sum = 0
for i in range(n):
    score = int(input("학생 "+ str(i+1) + "의 점수를 입력하세요"))
    sum = sum + score 

print("평균 점수 : ", sum/n)
```

```python
#예제 5 : while 문을 활용해 반 평균 수학 점수 구하기
n = int(input("학생수를 입력하세요!"))
i = 0
sum = 0
while i<n:
    score = int(input("학생 "+ str(i+1) + "의 점수를 입력하세요"))
    sum = sum + score 
    i = i+1
print("평균 점수 : ", sum/n)
```

### **반복문을 중간에 멈추고 싶다면? break를 활용하자!**
반복문으로 코드를 실행하던 중에 특정 조건에서 실행을 멈추거나 아예 종료하고 싶을 때가 있겠죠?  
이 때 break를 사용합니다! break가 있다면, 반복문의 조건과 상관없이 강제로 반복문을 중단시킵니다.  
입력한 값이 3의 배수이면 프로그램을 종료해보는 프로그램을 구현해봅시다!

```python
#예제 6: break 활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break
    print("n은 3의배수가 아닙니다!")
```

### **반복문을 중간까지만 실행하고 다시 처음으로 돌아오고 싶다면? continue를 활용하자!**
continue 또한 break와 마찬가지로 반복문 내에서 사용됩니다.  
break는 반복문을 강제로 중단한다면 continue는 반복문의 시작 부분으로 돌아가, for문의 변수를 진행하거나 while문의 반복조건을 판단합니다.   
이번에는 입력한 값이 3의 배수이면 종료하고, 4의 배수인 경우 4의 배수입니다를 출력하고, 3의배수가 아닙니다!를 출력하지 않게 구현해봅시다!

```python
#예제 6: continue 활용하기 
while True :
    n = int(input("숫자를 입력하세요!"))
    if n % 3 ==0:
        print("3의 배수가 입력되었습니다! 프로그램을 종료합니다")
        break
    if n % 4 ==0:
        print("4의 배수가 입력되었습니다! 처음으로 돌아갑니다.")
        continue
    print("n은 3의배수가 아닙니다!")
```
